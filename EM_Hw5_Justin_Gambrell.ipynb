```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
R=100
L=10
i=(-1)**(1/2)
w=np.arange(0,200)
y=[]
for j in w:
    vo_vi=(R**2-R*L*j*i)/(R**2+j**2*L**2)
    y.append(vo_vi)
plt.plot(w,y)
```

This is a low pass filter because it passes current at low frequency and does not pass current at high frequency
