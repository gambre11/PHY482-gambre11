```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

theta=np.arange(0,np.pi/2,.001)
theta
```

```python
ER_EI=[]
ET_EI=[]
for i in theta:
    ER=(((1-(.5*np.sin(i))**2)**(1/2))/np.cos(i)-2)/(((1-(.5*np.sin(i))**2)**(1/2))/np.cos(i)+2)
    ET=2/(((1-(.5*np.sin(i))**2)**(1/2))/np.cos(i)+2)
    ER_EI.append(ER)
    ET_EI.append(ET)
plt.plot(theta,ER_EI)
plt.plot(theta,ET_EI)
plt.xlabel('from 0 to pi/2')
names=['ER/EI','ET/EI']
plt.legend(names)
```

There is no angle that is not trivial that satisfies this equation.

```python
R=[]
T=[]
for i in theta:
    R1=((((1-(.5*np.sin(i))**2)**(1/2))/np.cos(i)-2)/(((1-(.5*np.sin(i))**2)**(1/2))/np.cos(i)+2))**2
    T1=2*(1-(.5*np.sin(i))**2)**(1/2)/np.cos(i)*(2/(((1-(.5*np.sin(i))**2)**(1/2))/np.cos(i)+2))**2
    R.append(R1)
    T.append(T1)
plt.plot(theta,R)
plt.plot(theta,T)
Rnew=np.array(R)
Tnew=np.array(T)
R_T=Rnew+Tnew
plt.plot(theta,R_T)
plt.xlabel('from 0 to pi/2')
names=['R','T','R+T']
plt.legend(names)
```

```python
#as a check
for i in R_T:
    if i==1 == False:
        print(i)
```

```python
for i in R_T:
    if i==1 == True:
        print(i)
```

This is simply a conservation of energy check. R+T will always equal 1 because energy is not created or destroyed, so the energy of the wave going in better be equal to the wave going through plus the wave reflected.
